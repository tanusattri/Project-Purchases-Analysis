```python
import sqlite3
import pandas as pd

# Connect to an in-memory SQLite database
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# Create a dummy Purchases table with some data
cursor.execute('''
CREATE TABLE IF NOT EXISTS Purchases (
    Purchase_ID INTEGER PRIMARY KEY,
    Course_name TEXT NOT NULL,
    Purchase_Date TEXT NOT NULL
);
''')

cursor.execute("INSERT INTO Purchases (Course_name, Purchase_Date) VALUES ('Python Basics', '2023-01-15')")
cursor.execute("INSERT INTO Purchases (Course_name, Purchase_Date) VALUES ('Data Science', '2023-01-20')")
cursor.execute("INSERT INTO Purchases (Course_name, Purchase_Date) VALUES ('Python Basics', '2023-02-01')")
cursor.execute("INSERT INTO Purchases (Course_name, Purchase_Date) VALUES ('Machine Learning', '2023-02-10')")
cursor.execute("INSERT INTO Purchases (Course_name, Purchase_Date) VALUES ('Data Science', '2023-01-20')")
cursor.execute("INSERT INTO Purchases (Course_name, Purchase_Date) VALUES ('Python Basics', '2023-03-05')")
cursor.execute("INSERT INTO Purchases (Course_name, Purchase_Date) VALUES ('Deep Learning', '2023-03-05')")
conn.commit()

# SQL Query 1: Find the highest purchased date
print("\n--- Highest Purchased Date ---")
query1 = """
SELECT STRFTIME('%Y-%m-%d', Purchase_Date) AS Highest_Purchased_Date,
       COUNT(*) AS Purchased_Count
FROM Purchases
GROUP BY Highest_Purchased_Date
ORDER BY Purchased_Count DESC
LIMIT 1;
"""
df1 = pd.read_sql_query(query1, conn)
print(df1)

# SQL Query 2: Find the month with the highest purchases
print("\n--- Month with Highest Purchases ---")
query2 = """
SELECT STRFTIME('%Y-%m', Purchase_Date) AS purchase_month,
       COUNT(*) AS purchase_count
FROM Purchases
GROUP BY purchase_month
ORDER BY purchase_count DESC
LIMIT 1;
"""
df2 = pd.read_sql_query(query2, conn)
print(df2)

# SQL Query 3: Find the first purchase date for each course
print("\n--- First Purchase Date for Each Course ---")
query3 = """
SELECT
    Course_name AS Course,
    MIN(Purchase_Date) AS First_Purchase_Date
FROM Purchases
GROUP BY Course
ORDER BY First_Purchase_Date ASC;
"""
df3 = pd.read_sql_query(query3, conn)
print(df3)

# Close the connection
conn.close()
```
